```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['0.1M', '0.2M', '0.5M', '0.8M', '1M']
treap = [1.71,3.86,11.89,20.64,26.1]
array = [0.09,0.18,0.41,0.71,0.82]
 
X_axis = np.arange(len(labels))
 
plt.bar(X_axis - 0.2, treap, 0.4, label = 'Treap')
plt.bar(X_axis + 0.2, array, 0.4, label = 'Dynamic Array')
plt.plot(X_axis - 0.2, treap, 0.4,linestyle='dashed',color = 'blue')
plt.plot(X_axis + 0.2, array, 0.4,linestyle='dashed',color = 'orange')
for a,b in zip(X_axis - 0.4, treap): 
    plt.text(a, b+0.4, str(b))
for a,b in zip(X_axis, array): 
    plt.text(a, b+0.4, str(b))
plt.xticks(X_axis, labels)
plt.xlabel("Lins")
plt.ylabel("Runtime")
plt.title("Exp1")
plt.legend()
plt.show()
```
